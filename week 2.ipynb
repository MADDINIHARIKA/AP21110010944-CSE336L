Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title

```python
import matplotlib.pyplot as plt
with open('data.txt', 'r') as file:
    data = [line.split() for line in file.readlines()]
    x_values = [float(row[0]) for row in data]
    y_values = [float(row[1]) for row in data]
plt.plot(x_values, y_values)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Line Plot from Data')
plt.show()
```

Create plot using subplots(), scatter, and some other methods on the Axes object

```python
import matplotlib.pyplot as plt
import numpy as np
x, y = np.random.rand(60), np.random.rand(60)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 1].plot(x, y, color='indigo', linestyle='-', marker='o', label='Line Plot')
axs[0, 0].scatter(x, y, color='pink', label='Scatter Plot')
bar_data = {'Category 1': 10, 'Category 2': 20, 'Category 3': 30}
axs[1, 0].bar(bar_data.keys(), bar_data.values(), color='pink', label='Bar Chart',edgecolor='black',width=0.6)
axs[1, 1].hist(np.random.randn(1000), bins=30, color='yellow', label='Histogram',edgecolor='black')
plt.show()
```

Generate random data for x and y and create a scatter plot using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y, color='pink', alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Random Data')
plt.show()
```

Create a bar chart to compare the sales of different products. You can use fictional data or come up
with your own dataset

```python
import matplotlib.pyplot as plt
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [350, 500, 280, 400] 
plt.bar(products, sales, color='pink')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Comparison of Sales by Product')
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

Generate a random dataset and create a histogram using Matplotlib to visualize the distribution of
the data

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0) 
data = np.random.randn(1000)
plt.hist(data, bins=30, color='black', edgecolor='pink')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.grid(True)
plt.show()
```

Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create
your own

```python
import matplotlib.pyplot as plt
categories = ['Housing', 'Food', 'Transportation', 'Utilities', 'Entertainment']
expenses = [2000, 1000, 800, 500, 300]
plt.figure(figsize=(8, 6))
plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Distribution of Expenses in Budget')
plt.axis('equal')
plt.show()
```
